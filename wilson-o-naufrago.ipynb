```python
#Importaremos as bibliotecas necessárias para o jogo rodar.

import random as rd
```

```python
#Criaremos a variável de tempo do jogo, dias e prob_sucess (variável que será usada para o sucesso de envio da carta). Talvez no futuro adicione mais ações.

action_list = ['write', 'send', 'drink', 'wait']
```

```python
"""
Definiremos uma classe chamada Person,
a qual possuirá atributos os quais representam condições iniciais de vida, sede, carta e contador de dias
de todos os objetos que sejam atribuídos com essa classe.
"""

class Person:
   
    def __init__(self):
        
        """
        Essa função é iniciada automaticamente ao ser chamado um objeto com essa classe.
        No jogo, estabelecerá as condições iniciais do objeto-personagem.
        """
        
        self.life = 100 #indicador de vida, sendo 0 o mínimo (= morte)
        self.letter = 0 #indicador de qualidade da carta escrita
        self.thirst = 0 #indicador de sede, sendo 0 o mínimo e 100 o máximo (= morte)
        self.day = 0 #indicador de dias passados
    
    def write(self):
        
        """
        Essa função será uma ação escolhida pelo jogador.
        Quando essa função é chamada, ela adiciona unidades na variável self.letter,
        como se o personagem estivesse investindo seu tempo no atribute self.letter.
        Por ser uma atividade, o personagem aumentará o indicador de sede e de dias.
        """
        
        self.letter += 10
        self.day += 1
        self.thirst += 10
        return print("Wilson has spent some time (1 day) writing.")
    
    def send(self):
        
        """
        Essa função será uma ação escolhida pelo jogador.
        Quando essa função é chamada, ela zera as unidades na variável self.letter, 
        como se o personagem enviasse a carta e portanto, fosse começar uma nova carta
        desse momento em diante.
        Por ser uma atividade, o personagem aumentará o indicador de sede e de dias.
        """

        self.day += 1
        self.thirst += 10
        return print("Wilson has spent some time (1 day) sending a letter.")
    
    def wait(self):
        
        """
        Essa função será uma ação escolhida pelo jogador.
        O personagem aumentará o indicador de sede e de dias, sem realizar nada relevante.
        """
        
        self.thirst += 10
        self.day += 1        
        return print("Wilson is lazy today, he won't do anything (1 day).")
    
    def drink(self):
        
        """
        Essa função será uma ação escolhida pelo jogador.
        Não realizada nada relevante.
        Quando essa função é chamada, ela pula o dia.
        """
        
        self.thirst += -30
        self.day += 1
        return print("Wilson drank some water (1 day).")
        
    def snake(self):
        
        """
        Essa função será um evento aleatório que ocorre após a passagem dos dias.
        """
        
        prob_snake = rd.random()
        if prob_snake < 0.3:
            self.life += -20
            return print("A snake has bitten Wilson!")
        else:
            return print("Wilson is safe.")
```

```python
print("""

|  \  _  |  \|  \|  \                                                                                                   
| $$ / \ | $$ \$$| $$  _______   ______   _______                                                                       
| $$/  $\| $$|  \| $$ /       \ /      \ |       \                                                                      
| $$  $$$\ $$| $$| $$|  $$$$$$$|  $$$$$$\| $$$$$$$\                                                                     
| $$ $$\$$\$$| $$| $$ \$$    \ | $$  | $$| $$  | $$                                                                     
| $$$$  \$$$$| $$| $$ _\$$$$$$\| $$__/ $$| $$  | $$ __                                                                  
| $$$    \$$$| $$| $$|       $$ \$$    $$| $$  | $$|  \                                                                 
 \$$      \$$ \$$ \$$ \$$$$$$$   \$$$$$$  \$$   \$$| $$                                                                 
   __      __                         ______        \$               __                                                 
  |  \    |  \                       /      \                       |  \                                                
 _| $$_   | $$____    ______        |  $$$$$$\  ______    _______  _| $$_     ______   __   __   __   ______   __    __ 
|   $$ \  | $$    \  /      \       | $$   \$$ |      \  /       \|   $$ \   |      \ |  \ |  \ |  \ |      \ |  \  |  \
 \$$$$$$  | $$$$$$$\|  $$$$$$\      | $$        \$$$$$$\|  $$$$$$$ \$$$$$$    \$$$$$$\| $$ | $$ | $$  \$$$$$$\| $$  | $$
  | $$ __ | $$  | $$| $$    $$      | $$   __  /      $$ \$$    \   | $$ __  /      $$| $$ | $$ | $$ /      $$| $$  | $$
  | $$|  \| $$  | $$| $$$$$$$$      | $$__/  \|  $$$$$$$ _\$$$$$$\  | $$|  \|  $$$$$$$| $$_/ $$_/ $$|  $$$$$$$| $$__/ $$
   \$$  $$| $$  | $$ \$$     \       \$$    $$ \$$    $$|       $$   \$$  $$ \$$    $$ \$$   $$   $$ \$$    $$ \$$    $$
    \$$$$  \$$   \$$  \$$$$$$$        \$$$$$$   \$$$$$$$ \$$$$$$$     \$$$$   \$$$$$$$  \$$$$$\$$$$   \$$$$$$$ _\$$$$$$$
                                                                                                              |  \__| $$
                                                                                                               \$$    $$
                                                                                                                \$$$$$$ 


PORTUGUES: Wilson, o náufrago
Wilson decidiu beber vinho e manejar seu avião. Má idéia!
Acabou caindo em uma ilha deserta na costa paulista cheia de cobras!
Agora ele precisa avisar os seus parentes do seu paradeiro. Acontece que Wilson não tem uma boa relação com a sua família
e necessita convencê-los a virem buscá-lo.
Ajude Wilson a escrever cartas convincentes e enviá-las à sua família, através de suas garrafas de vinho.
Mas lembre-se de manter o nosso amigo hidratado e longe das cobras!

ENGLISH: Wilson, the castway
Wilson decided to drink wine and drive his airpline. Bad idea!
He ended up crashing his plane on a desert island, full of snakes! Yikes!
Now he needs to send a message to his family of his whereabouts.
Since Wilson does not have a very heartwarming relationship with his relatives, he has to convince his family
to come pick him up. Help him write letters and send them inside his wine bottles.
But remember to keep our friend hydrated and away from the snakes!
""")


def game():
    
    """
    Essa função conterá o loop While que dita o jogo e as condições de vitória/fracasso.
    Definiremos o nosso personagem Wilson, que terá os mesmos atributos e métodos da classe Person.
    Definiremos o tempo de parada do jogo com base nas três condições:
    - Ou o Wilson consegue enviar a carta (prob_sucess > 80%).
    - Ou o Wilson chega a 0 ou menos pontos de vida.
    - Ou o Wilson chega a 100 ou mais de sede.
    Se alguma dessas condições é satisfeita, o loop While termina.
    """
    
    wilson = Person()
    win = False
    
    while wilson.life > 0 and win == False and wilson.thirst < 100:

        # Essas são as descrições do nível da carta.
        if wilson.letter <= 20:
            letter_description = 'it is unreadable - even a monkey could write something better!'
        elif wilson.letter <= 40:
            letter_description = 'there are some words written down but nothing really interesting.'
        elif wilson.letter <= 60:
            letter_description = 'there is a text, fairly elaborated. This is becoming promising!'
        elif wilson.letter <= 80:
            letter_description = 'there is an amazing poem written by out castway! This will surely touch his relatives hearts!'
        elif wilson.letter <= 100:
            letter_description = 'it is a masterpiece - our castaway is a genius! Even Shakespeare would be jealous of him. What are you waiting for to send it?'
        
        # Essas são as descrições do nível de vida.
        if wilson.life >= 100:
            life_description = 'Wilson has never been in better shape.'
        elif wilson.life >= 80:
            life_description = 'Wilson has been feeling a bit weird.'
        elif wilson.life >= 60:
            life_description = 'Wilson is not feeling well, he cant barely get out of his bed.'
        elif wilson.life >= 40:
            life_description = 'Wilson is seriously ill and we better get the hell out of this island!'
        elif wilson.life >= 20:
            life_description = 'Wilson is having a near-death experience and he is seeing an oasis in the middle of the island!'
        
        # Essas são as descrições do nível de sede.
        if wilson.thirst <= 20:
            thirst_description = 'Wilson is not thirsty at all, what a strong man!'
        elif wilson.letter <= 40:
            thirst_description = 'Wilson would like some coconut water for sure.'
        elif wilson.letter <= 60:
            thirst_description = 'Wilson is thirsty, he needs some fresh water right now. '
        elif wilson.letter <= 80:
            thirst_description = 'Wilson is nearly dying of thirst, you better find something for him to drink!'
        elif wilson.letter <= 100:
            thirst_description = 'Wilson is dying!'        
        
        
        # Essa é a mensagem que aparece no início do dia para o usuário.
        action_general = input(f"\n"
                       f"Today is day {wilson.day}. Currently...\n"
                       f"Wilson has {wilson.life}% life which means {life_description}\n"
                       f"Wilson has {wilson.thirst}% of thirst which means {thirst_description}\n"
                       f"Wilson has a letter at {wilson.letter}% completion which means {letter_description}\n"
                       f"What will Wilson do today? {action_list}.")

        # Vamos permitir que o usuário coloque palavras em letras maiúsculas
        action = action_general.lower()
        
        # Essas são as ações tomadas pelo usuário e a chamada dos métodos da classe Person para o personagem:
        if action == 'write': #utilizaremos os métodos referentes a escrever e ao ataque de cobra
            wilson.write()
            wilson.snake()
        elif action == 'send': #utilizaremos os métodos referentes a enviar e ao ataque de cobra
            wilson.send()                
            if wilson.letter > 80: #utilizaremos dentro do código o IF para ativar o flag 'win', já que essa variable não consegue ser alterada dentro do método da classe (é uma variável local de game)
                win = True
            else:
                pass
            wilson.snake()
            wilson.letter = 0 #zeramos o contado, simulando o envio da carta acumulada
        elif action == 'wait': #utilizaremos os métodos referentes a esperar e ao ataque de cobra
            wilson.wait()
            wilson.snake()
        elif action == 'drink':
            wilson.drink()
            if wilson.thirst <0: #para evitar sede abaixo de zero, seria um pouco absurdo
                wilson.thirst = 0
            wilson.snake()     
        else: #caso o usuário insira qualquer outra coisa que não esteja na lista action_list
            print("Wilson doesn't know what you want, write something from the action list.")
    
            
    #No caso de alguma das condições impostas no loop While serem violadas, ocorre o fim do jogo:       

    if win == True:
        print("One of Wilson's letters has reached his family! You win!")
    elif wilson.life <= 0:
        print("Wilson was killed by a snake! You lose!")
    elif wilson.thirst >= 100:
        print("Wilson died of thirst! You lose!")
    else:
        print("You crashed the game!")

#Variável again funciona como flag, o primeiro jogo já inicia automaticamente.        
again = 'yes' 

#Para iniciar um novo jogo ao final do outro, inserimos esse loop While que muda o flag para 'no' se o jogador inserir essa resposta, parando o jogo.
while again == 'yes':
    game()
    again = input("Would you like to play again? yes or no?") #Se o usuário responde 'No', o while para e o jogo não reinicia.
```
