```python
#Importaremos as bibliotecas necessárias para o jogo rodar.

import random as rd
```

```python
#Criaremos a variável de tempo do jogo, dias e prob_sucess (variável que será usada para o sucesso de envio da carta).

prob_sucess = 0
action_list = ['write', 'send', 'wait']
```

```python
"""
Definiremos uma classe chamada Person,
a qual possuirá atributos os quais representam condições iniciais de vida, sede e carta
de todos os objetos que sejam atribuídos com essa classe.
"""

class Person:
   
    def __init__(self):
        
        """
        Essa função é iniciada automaticamente ao ser chamado um objeto com essa classe.
        No jogo, estabelecerá as condições iniciais do objeto-personagem.
        """
        
        self.life = 100
        self.letter = 0
        self.day = 0
    
    def write(self):
        
        """
        Essa função será uma ação escolhida pelo jogador.
        Quando essa função é chamada, ela adiciona 0.1 unidades na variável self.letter e
        passa um dia, como se o personagem estivesse investindo seu tempo no atribute self.letter.
        """
        
        self.letter += 0.1
        self.day += 1
        return print("Wilson has spent some time (1 day) writing.")
    
    def send(self):
        
        """
        Essa função será uma ação escolhida pelo jogador.
        Quando essa função é chamada, ela zera as unidades na variável self.letter e
        passa um dia, como se o personagem enviasse a carta e portanto, fosse começar uma nova carta
        desse momento em diante.
        
        Também alterará o valor de rd.random() base do jogo, e poderá ocasionar fim do While caso o rd.random()
        atinja valor maior que 0.8.
        """
                  
        prob_sucess = rd.random() + self.letter
        self.letter = 0
        self.day += 1
        return print("Wilson has spent some time (1 day) sending a letter.")
    
    def wait(self):
        
        """
        Essa função será uma ação escolhida pelo jogador.
        Quando essa função é chamada, ela pula o dia.
        """
        
        self.day += 1        
        return print("Wilson is lazy today (1 day)")
        
    def snake(self):
        
        """
        Essa função será um evento aleatório que ocorre após a passagem dos dias.
        """
        
        prob_snake = rd.random()
        if prob_snake < 0.3:
            self.life += -20
            return print("A snake has bitten our friend Wilson!")
        else:
            return print("Wilson is safe.")
```

```python
"""
Definiremos o nosso personagem Wilson, que terá os mesmos atributos e métodos da classe Person.
"""

"""
Definiremos o tempo de parada do jogo com base nas três condições:
- Ou o Wilson consegue enviar a carta (prob_sucess > 0.8).
- Ou o Wilson chega a 0 ou menos pontos de vida.
"""

def game():
    
    wilson = Person()
    
    while prob_sucess < 0.8 and wilson.life > 0:
    
        action = input(f"Today is day {wilson.day}. What will Wilson do today? {action_list}.")
        
        if action == 'write': #utilizaremos os métodos referentes a escrever e ao ataque de cobra
            wilson.write()
            wilson.snake()
        elif action == 'send': #utilizaremos os métodos referentes a enviar e ao ataque de cobra
            wilson.send()
            wilson.snake()
        elif action == 'wait': #utilizaremos os métodos referentes a esperar e ao ataque de cobra
            wilson.wait()
            wilson.snake()
        else: #caso o usuário insira qualquer outra coisa que não esteja na lista action_list
            print("Wilson doesn't know what you want, write something from the action list.")
    
            
    #No caso de alguma das condições impostas no loop While serem violadas, ocorre o fim do jogo:       
            
    if prob_sucess >= 0.8:
        print("One of Wilson's letters has reached his family! You win!")
    elif wilson.life <= 0:
        print("Wilson is dead! You lose!")
    else:
        print("You crashed the game!")

again = 'Yes' #Variável again funciona como flag, o primeiro jogo já inicia automaticamente.

while again == 'Yes': #While para iniciar um novo jogo.
    game()
    again = input("Would you like to play again? Yes or No.") #Se o usuário responde 'No', o while para e o jogo não reinicia.
```
